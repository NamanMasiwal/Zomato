# Dropping Duplicates

```python
df.drop_duplicates(inplace = True )
df.shape
```

```python
df['rate'].unique()
```

```python
def handlerate(value):
    if(value== 'NEW' or value == '-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
        
```

```python
df.rate.isnull().sum()
```

# Filling Nul Values in Rate Column With Mean

```python
df['rate'].fillna(df['rate'].mean(),inplace = True)
df['rate'].isnull().sum
```

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',' , '')
        return float(value)
    else:
        return float(value)

df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

# Cleaning Rest type COlumn

```python
df['rest_type'].value_counts()
```

```python
rest_type = df['rest_type'].value_counts(ascending = False)
rest_type
```

```python
rest_type_lessthan1000 = rest_type[rest_type<1000]
rest_type_lessthan1000
```

# Making Rest Type less than 1000 in Frequency as other

```python
def handl_rest_type(value):
    if(value in rest_type_lessthan1000):
        return 'others'
    else:
        return value
df['rest_type']= df['rest_type'].apply(handl_rest_type)
df['rest_type'].value_counts()
```

# Cleaning Location Column

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending = False)
location_lessthan300 = location[location<300]

def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
    
df['location']=df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

```python
cuisines = df['cuisines'].value_counts(ascending = False)

cuisines_less than 100 = cuisines [cuisines<100]

def handle_cuisines (value):

    if(value in cuisines_lessthan100):

        return 'others'
    else:
    return value
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

# No.of Votes,Location Wise

```python
# Selecting 'location' and 'votes' columns from the DataFrame
df4 = df[['location', 'votes']]

# Dropping duplicate rows
df4 = df4.drop_duplicates()

# Grouping by 'location' and summing the 'votes' column
df5 = df4.groupby(["location"])['votes'].sum()

# Converting the result to a DataFrame
df5 = df5.to_frame()

# Sorting the DataFrame by the 'votes' column in descending order
df5 = df5.sort_values('votes', ascending=False)

# Displaying the top entries
df5.head()
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(x=df5.index, y=df5['votes'])
plt.xticks(rotation=90)
plt.xlabel('Location')
plt.ylabel('Total Votes')
plt.title('Total Votes by Location')
plt.show()
```
